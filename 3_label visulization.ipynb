```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import glob
import os
```

```python
base_path = r'D:\TUM论文工作\第五周工作\数据\数据5'
files = os.listdir(base_path)
files.sort(key=lambda x: int(x.split('_')[0]))
dfs1=[]
for path in files:
    full_path = os.path.join(base_path, path)
    dfs1.append(pd.read_csv(full_path))
```

```python
data_number=1
a=[10]
num_lanechange=1
```

```python
j=0
for k in range(data_number):
    df1=dfs1[k]
    df1=df1[df1['numLaneChanges']==num_lanechange]
    #df1=df1[df1['numLaneChanges']>num_lanechange]
    c=df1['id']
    d=c.drop_duplicates()
    for i in d:
        dfid=df1[df1['id']==i]
        b=dfid.shape
        dfid.index=np.arange(b[0])
        plt.plot(dfid.loc[:,'x'],dfid.loc[:,'y'])
        plt.plot(dfid.loc[:,'x'],dfid.loc[:,'laneId'])
        plt.plot(dfid.loc[:,'x'],dfid.loc[:,'yVelocity']*10)
        plt.plot(dfid.loc[:,'x'],np.full((dfid.shape[0]),2),'k--')
        plt.plot(dfid.loc[:,'x'],dfid.loc[:,'label'])
        plt.title('id='+str(i)+'/'+str(dfid.loc[1,'drivingDirection']))
        plt.legend()
        plt.savefig(r"D:\TUM论文工作\第五周工作\数据\图片1\%d_%d.jpg"%(a[j],i))
        #plt.savefig(r"D:\TUM论文工作\第五周工作\数据\图片2\%d_%d.jpg"%(a[j],i))
        plt.show()
        plt.close()
j=j+1
```
